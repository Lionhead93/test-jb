# 문제: Two Sum
정수 배열 nums와 정수 target이 주어졌을 때,
두 수를 더해 target을 만들 수 있는 서로 다른 인덱스 두 개를 반환하라.

예시:
- 입력: nums = [2, 7, 11, 15], target = 9
- 출력: [0, 1]

```
nums = [2, 7, 11, 15]
target = 9
```

## 방법 1: 브루트포스 (O(n²))

```
def two_sum_bruteforce(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]

two_sum_bruteforce(nums, target)
```

## 방법 2: 해시맵 사용 (O(n)) — 가장 효율적

```
def two_sum_hash(nums, target):
    hashmap = {}
    for i, num in enumerate(nums):
        need = target - num
        if need in hashmap:
            return [hashmap[need], i]
        hashmap[num] = i

two_sum_hash(nums, target)
```

```
result = two_sum_hash(nums, target)
print('결과:', result)
```

## 추가 테스트

```
tests = [
    ([3, 2, 4], 6),
    ([3, 3], 6),
    ([1, 5, 3, 7], 8),
]

for nums, target in tests:
    print(nums, target, "->", two_sum_hash(nums, target))
```
